```python
# zad1 

all_cases=[]
by_date={}
by_country={}
def load_data():
    with open('C:\\Users\\K\\Desktop\\studia\\skryptLab\\Covid.txt', 'r',encoding="utf-8") as file:
        for line in file:
            row = line.strip().split()
            row[3] = int(row[3])
            row[2] = int(row[2])
            row[1] = int(row[1])
            row[5] = int(row[5])
            row[4] = int(row[4])
            all_cases.append((row[6],row[3],row[2],row[1],row[5],row[4])) 

            if((row[3],row[2],row[1]) in  by_date):
                by_date[(row[3],row[2],row[1])].append((row[6],row[5],row[4]))
            else:
                by_date[(row[3],row[2],row[1])]=[(row[6],row[5],row[4])]  

            if(row[6] in  by_country):
                by_country[row[6]].append((row[3],row[2],row[1],row[5],row[4]))
            else:
                by_country[row[6]]=[(row[3],row[2],row[1],row[5],row[4])]

    file.close()

```

```python
# zad 2
from timeit import default_timer as timer

#   ALL_CASES
def for_date_a(year, month, day):
    start = timer()   
    date = (year,month,day)
    
    deathTotal = 0
    casesTotal = 0
    
    for (_,y,m,d,deaths,cases) in all_cases:
        if ((y,m,d) == date):   
            deathTotal+=deaths
            casesTotal+=cases
    end=timer()
    execTime = round((end - start)*1000,2)
    
    print("   Na podstawie all_cases")
    print("   Łączna liczba przypadków: {} \n   Łączna liczba zgonów: {}".format(casesTotal,deathTotal))
    print("   Czas wykonania funkcji: {}".format(execTime))

# BY_DATE
def for_date_d(year, month, day):
    start = timer()   
    date = (year,month,day)
    
    deathTotal = 0
    casesTotal = 0
    
    for (_,deaths,cases) in by_date[date]:
        deathTotal+=deaths
        casesTotal+=cases
    
    end=timer()
    execTime = round((end - start)*1000,2)
    
    
    print("   Na podstawie by_date")
    print("   Łączna liczba przypadków: {} \n   Łączna liczba zgonów: {}".format(casesTotal,deathTotal))
    print("   Czas wykonania funkcji: {}".format(execTime))
        
# BY_COUNTRY
def for_date_c(year, month, day):
    start = timer()   
    date = (year,month,day)
    
    deathTotal = 0
    casesTotal = 0
    
    for val in by_country:
        for (y,m,d,deaths,cases) in by_country[val]:
            if ((y,m,d)==date):
                deathTotal+=deaths
                casesTotal+=cases
    
    end=timer()
    execTime = round((end - start)*1000,2)
    
    print("   Na podstawie by_country")
    print("   Łączna liczba przypadków: {} \n   Łączna liczba zgonów: {}".format(casesTotal,deathTotal))
    print("   Czas wykonania funkcji: {}".format(execTime))
    
```
