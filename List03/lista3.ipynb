```python
# zadanie 1 
class Controlled_text:
    
    def __init__(self):
        self._text=None
    
    def __lt__(self,other):
        return self.text<other.text
    
    def __gt__(self,other):
        return self.text>other.text
    
    def __eq__(self,other):
        return self.text==other.text
    
    def __le__(self,other):
        return self.text<=other.text
    
    def __ge__(self,other):
        return self.text>=other.text
    

    
    @property
    def text(self):
        #print("Text getter called")
        
        return self._text
    
    @text.setter
    def text(self, text):
        
        if (len(text) < 1):
            raise ValueError("Given value must be at least 1 char")
            
        elif (' ' in text):
            raise ValueError("Given value must not contain spaces")
            
        elif (not text.isprintable()):
            raise ValueError("Given value is not printable")
        
        #print("Text setter called")
        
        self._text=text
```

```python
# zadanie 2

# loading names from file
def load_names():
    file1 = open('C:\\Users\\K\\Downloads\\PopularneImiona.txt', 'r', encoding="utf8")
    names=[]

    for line in file1.readlines():
        
        if (not line.isspace()):
            name = line.split()[0]
            names.append(name)
    names.remove('Kobiety')
    names.remove('Mężczyźni') 
    return names
        
class First_name(Controlled_text):   #‪ C:\Users\K\Downloads\PopularneImiona.txt
    
    validNames = load_names()   # poprawne(np.Zofia, Julia, Jakub, Leon)
                                # niepoprawne(np. Monika, Patryk)
    
    def __init__(self,name):
        super().__init__()
        self.name=name
   
    @property
    def name(self):     
        return self._name
    
    @name.setter
    def name(self, name):
        tempText=self.text  # przechowanie self.text 
        
        self.text=name   # sprawdzanie czy name jest prawidlowe(co najmniej jeden znak, brak spacji itd.)
        
        self._text = tempText # przypisane self._text poprzedniej wartosci
        
        name = name.capitalize()     
        self.check_name(name)       
        self._name=name

        
    def check_sex(self,name):  # sprawdzanie plci
        if (self.check_name(name)):
            if(name.endswith('a')):
                return True
            return False

     
    def is_male(self):
        return not self.check_sex(self._name)
        
    def is_female(self):
        return self.check_sex(self._name)
    
    def male_name(self,name):   
        return not self.check_sex(name)
 
    def female_name(self,name):      
        return self.check_sex(name)
    
        
    def check_name(self,name):   # sprawdzanie czy name jest w PoprawneImiona.txt
        if (not name.capitalize() in First_name.validNames):
            raise ValueError("{} is not in valid names list".format(name))
        return True
```

```python
# zadanie 3
class Last_name(Controlled_text):

    def __init__(self,lname):
        super().__init__()
        self.lname=lname
   
    @property
    def lname(self):     
        return self._lname
    
    @lname.setter
    def lname(self, lname):
        tempText=self.text  # przechowanie self.text 
        
        self.text=lname   # sprawdzanie czy lname jest prawidlowe(co najmniej jeden znak, brak spacji itd.)
        
        self._text = tempText # przypisane self._text poprzedniej wartosci
        
        lname=lname.split('-') # lista zlozona z czlonow nazwiska
        
        if(len(lname)<=2):      # nazwisko moze byc maksymalnie dwuczlonowe
            for name in lname:   # przejscie po liscie czlonow
                if (not name==name.capitalize()):   #sprawdzenie czy czlon jest poprawnie napisany
                    raise ValueError("Last name is not valid")
            self._lname='-'.join(lname) # zlaczenie nazwiska
        
        else:
            raise ValueError("Last name must be maksymalnie dwuczłonowe") 
            
```
